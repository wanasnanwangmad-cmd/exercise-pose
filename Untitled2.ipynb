```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt

# Input / Output
dirtiness = ctrl.Antecedent(np.arange(0, 11, 0.1), 'dirtiness')
load = ctrl.Antecedent(np.arange(0, 11, 0.1), 'load')
wash_time = ctrl.Consequent(np.arange(0, 91, 1), 'wash_time')

# Membership Functions
dirtiness['low'] = fuzz.trimf(dirtiness.universe, [0, 0, 4])
dirtiness['medium'] = fuzz.trimf(dirtiness.universe, [2, 4, 6])
dirtiness['high'] = fuzz.trimf(dirtiness.universe, [5, 7.5, 10])

load['small'] = fuzz.trimf(load.universe, [0, 2.5, 5])
load['large'] = fuzz.trimf(load.universe, [4, 7, 10])

wash_time['short'] = fuzz.trimf(wash_time.universe, [0, 15, 30])
wash_time['medium'] = fuzz.trimf(wash_time.universe, [20, 35, 50])
wash_time['long'] = fuzz.trimf(wash_time.universe, [40, 65, 90])

# Rules
rule1 = ctrl.Rule(dirtiness['low'] & load['small'], wash_time['short'])
rule2 = ctrl.Rule(dirtiness['medium'] & load['small'], wash_time['medium'])
rule3 = ctrl.Rule(dirtiness['high'] & load['small'], wash_time['long'])
rule4 = ctrl.Rule(dirtiness['low'] & load['large'], wash_time['medium'])
rule5 = ctrl.Rule(dirtiness['medium'] & load['large'], wash_time['long'])
rule6 = ctrl.Rule(dirtiness['high'] & load['large'], wash_time['long'])

# Control System
washing_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6])
washing = ctrl.ControlSystemSimulation(washing_ctrl)

# Test input
washing.input['dirtiness'] = 7
washing.input['load'] = 5
washing.compute()

print("Fuzzy output (wash time):", washing.output['wash_time'], "minutes")

# Show graphs
dirtiness.view(sim=washing)
load.view(sim=washing)
wash_time.view(sim=washing)
plt.show()
```
