```python

triTypes = ["", "scalene", "isosceles", "equilateral", "not a valid triangle"]

a = int(input("Enter number A: "))
b = int(input("Enter number B: "))
c = int(input("Enter number C: "))

if (a < b + c) and (b < a + c) and (c < b + a):
    if (a == b) and (b == c):
        triOut = 3
    elif (a != b) and (a != c) and (b != c):
        triOut = 2
    else:
        triOut = 4
else:
    triOut = 1

print(triTypes[triOut])
```
